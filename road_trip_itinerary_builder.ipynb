```python
import requests
from datetime import datetime, timedelta
import getpass

weather_api_url = "https://api.shecodes.io/weather/v1/current"
ai_api_url = "https://api.shecodes.io/ai/v1/generate"

weather_api_key = getpass.getpass(" API key: ").strip()
```

```python
def get_distance(origin, destination):
    prompt = f"What is the driving distance between {origin} and {destination} in miles and kilometers?"
    params = {
        "prompt": prompt,
        "context": "road trip distance calculator",
        "key": weather_api_key
    }
    response = requests.get(ai_api_url, params=params)
    if response.status_code == 200:
        return response.json().get("answer", "Distance unavailable")
    return "Distance unavailable"


def get_weather(city):
    params = {"query": city, "key": weather_api_key}
    response = requests.get(weather_api_url, params=params)
    if response.status_code == 200:
        data = response.json()
        temp = data.get("temperature", {}).get("current")
        condition = data.get("condition", {}).get("description", "")
        return f"{temp}Â°C, {condition}"
    return "Weather data unavailable"


def get_daily_plan(city, day_num):
    prompt = (
        f"Create a road trip itinerary for day {day_num} in {city}. "
        f"List exactly two bullet points using emojis: "
        f"ğŸ½ï¸ A place to eat with a cost in $00.00 format. "
        f"ğŸ“A place to visit or explore with a cost in $00.00 format. "
        f"Only use emojis for the list items. No intro or summary. No bold or markdown symbols."
    )
    params = {
        "prompt": prompt,
        "context": "fun and budgeted travel itinerary",
        "key": weather_api_key
    }
    response = requests.get(ai_api_url, params=params)
    if response.status_code == 200:
        return response.json().get("answer", "No suggestions for this day.")
    return "Error generating itinerary."
```

```python
print("ğŸš— Welcome to the Road Trip Itinerary Builder!")
start_city = input("Where are you starting your road trip from? ").strip()
destination = input("Where are you heading to? ").strip()
num_days = int(input("How many days will you be staying there? "))

print("\n" + "=" * 60)
print(f"# ğŸ§­ Road Trip Itinerary: {start_city} to {destination}")
print(f"ğŸ›£ï¸ Distance: {get_distance(start_city, destination)}")
print(f"ğŸŒ¤ï¸ Current weather in {start_city}: {get_weather(start_city)}")
print(f"ğŸŒ¤ï¸ Current weather in {destination}: {get_weather(destination)}")
print("=" * 60 + "\n")

for day in range(num_days):
    date_str = (datetime.now() + timedelta(days=day)).strftime("%A, %B %d")
    weather = get_weather(destination)
    plan = get_daily_plan(destination, day + 1)

    print(f"## Day {day + 1}: {date_str}")
    print(f"ğŸŒ¡ï¸ Weather: {weather}")
    print("ğŸ—ºï¸ Plan:")
    for line in plan.strip().splitlines():
        print(f"- {line.strip()}")
    print("\n" + "-" * 50 + "\n")

print("ğŸ‰ Thank you for using the Road Trip Itinerary Builder!")
print("âœ¨ Built with ğŸ’» + â˜• by Jessica Flores. Safe travels! ğŸ›£ï¸âœ¨")
```
